# 30. 形態素解析結果の読み込み

形態素解析結果（neko.txt.mecab）を読み込むプログラムを実装せよ．ただし，各形態素は表層形（surface），基本形（base），品詞（pos），品詞細分類1（pos1）をキーとするマッピング型に格納し，1文を形態素（マッピング型）のリストとして表現せよ．第4章の残りの問題では，ここで作ったプログラムを活用せよ．

```
# 読み込むファイルのパス
path = '/content/drive/My Drive/Colab Notebooks/NLP-100/data/'
file_name = 'neko.txt.mecab'

# ファイルの読み込み
data = ''
with open(path + file_name) as f:
  data = f.read()

# 読み込み
text_list = []
for line in data.split('\n'):
  # タブをカンマに変換
  line = line.replace('\t', ',')
  # カンマ区切りのリスト
  line_list = line.split(',')

  if line_list[0] != 'EOS':
    morpheme_dic = {}
    morpheme_dic['surface'] = line_list[0]
    morpheme_dic['base'] = line_list[7]
    morpheme_dic['pos'] = line_list[1]
    morpheme_dic['pos1'] = line_list[2]
    text_list.append(morpheme_dic)

text_list
```

# 31. 動詞

動詞の表層形をすべて抽出せよ．

```
surface_of_verbs = []

for morpheme_dic in text_list:
  if morpheme_dic['pos'] == '動詞':
    surface_of_verbs.append(morpheme_dic['surface'])

print(surface_of_verbs)
```

# 32. 動詞の原形

動詞の原形をすべて抽出せよ．

```
verb_stems = []

for morpheme_dic in text_list:
  if morpheme_dic['pos'] == '動詞':
    verb_stems.append(morpheme_dic['base'])

print(verb_stems)
```

# 33. 「AのB」

2つの名詞が「の」で連結されている名詞句を抽出せよ．

```
nouns = []

for index, morpheme_dic in enumerate(text_list):
  if morpheme_dic['pos'] == '名詞' and text_list[index+1]['surface'] == 'の' and text_list[index+2]['pos'] == '名詞':
    noun = morpheme_dic['surface'] + text_list[index+1]['surface'] + text_list[index+2]['surface']
    nouns.append(noun)
  
print(nouns)
```

# 34. 名詞の連接

名詞の連接（連続して出現する名詞）を最長一致で抽出せよ

```
concatenated_nouns = []

scanned_index = []

for index, morpheme_dic in enumerate(text_list):
  if not index in scanned_index and morpheme_dic['pos'] == '名詞' and text_list[index+1]['pos'] == '名詞':
    scanned_index.append(index)
    tmp = []
    tmp.append(morpheme_dic['surface'])
    j = index + 1

    while  text_list[j]['pos'] == '名詞':
      scanned_index.append(j)
      tmp.append(text_list[j]['surface'])
      j = j + 1
    
    concatenated_nouns.append(''.join(tmp))

print(concatenated_nouns)
```
